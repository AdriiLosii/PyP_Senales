```python
import matplotlib.pyplot as plt
from IPython.display import display
import sympy as sp
from sympy.abc import t


import sympy as sp
from IPython.display import display

sp.init_printing(use_unicode=True)

# Usaremos sp.fourier_series
# fourier_series(f, (varuabkem lim_inf, lim_sup))


t = sp.symbols("t")

def funcion_dientes_sierra(t):
    return t

def funcion_raruna(t):
    return sp.Piecewise((1, (t>=-sp.pi)&(t<0)), (0, (t>=0)&(t<sp.pi/2)), (0.5, (t>=sp.pi/2)&(t<sp.pi)), (0, True))

limites = (t, -sp.pi, sp.pi)
salida = sp.fourier_series(funcion_dientes_sierra(t), limites)

display(salida)
display(salida.truncate(n=10))

p = sp.plot(funcion_dientes_sierra(t),salida.truncate(n=5),salida.truncate(n=10),salida.truncate(n=30),(t,-10, 10), show=False, legend=False, title="Fourier Series. Sawtooth wave ")
#p = sp.plot(funcion_raruna(t),salida.truncate(n=5),salida.truncate(n=10),salida.truncate(n=30),(t,-10, 10), show=False, legend=False, title="Fourier Series. Sawtooth wave ")

print("###Start Info plot object###")
print(p)
print("###End Info plot object###")
p[0].line_color="black"
p[1].line_color="blue"
p[2].line_color="red"
p[3].line_color="green"
p.show()
```

```python
display(salida.a0)
display(salida.an)
display(salida.bn)

# coeficiente 7
display(salida.bn.coeff(7))

# maximo coeficiente 4
display(salida.bn.coeff(4).subs(t, sp.pi/8.0))
```
