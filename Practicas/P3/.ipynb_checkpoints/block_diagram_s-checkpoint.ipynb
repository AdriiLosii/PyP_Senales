# Diagramas de bloques
* Representan circuitos para el análisis y síntesis de las señales
*  Nos permiten diseñar de forma abstracta algoritmos sin importar como serán implementados
* Se realizará empleando  bloques de construcción básicos que representan las operaciones más elementales sobre las señales
* El funcionamiento resumido es el siguiente:
 * El circuitor/sistema recibe una secuencia de entrada $x[n]$
 * Se aplican a las operaciones definidas en los bloques de construcción que componen el circuito
 * El circuito/sistema devuelve una secuencia de salida $y[n]$


```python
#librerias necesarias para la práctica
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from random import randrange
import IPython
plt.rcParams["figure.figsize"] = (14,4) #dimensiones por defecto al plotear (width, height)

#Este es un método de visualización muy básico para hacer pruebas
def plot_stem( t , y, y_lim=None, view_baseline=True, title=None):
    """Función que configura y visualiza un gráfico stem 

       Parámetros:
       t -- muestras de tiempo
       y -- secuencia para visualizar
       y_lim -- tupla de la forma (min, max) que establece los límites en el eje de ordenadas. Por defecto None
       view_baseline -- Establece una línea horizontal en el eje de las x a la altura de  y=0 . Por defecto False

    """
    markerline, stemlines, baseline = plt.stem(t, y)
    plt.setp(baseline, visible=view_baseline)
    plt.setp(markerline, color="crimson",markersize = 10)
    plt.setp(stemlines, color="crimson", linewidth = 4)
    plt.axvline(x=0, linewidth=1,color="black")
    plt.ylim(y_lim)
    if title is not None:
        plt.title(title)
    plt.show()
```

## Implementación de bloques básicos
Las operaciones básicas que vamos a implementar iniciales son:

* Sumador de señales
* Multiplicador de señales
* Escala (multiplicador por una constante)
* Retardador genérico

En esta primera aproximación no se pretende que se haga una implementación óptima, si no una optimización funcional pero no está limitado a una solución concreta. Si se quiere tener una librería funcional, óptima y expandible entonces requiere un poco más de análisis.

```python
#
# objeto señal: almacenamos datos y muestras
#
class sennal():
    def __init__(self, d, t):
        if len(d) != len(t): raise Exception("oops! data samples and time have different sizes")
        self.d = np.array(d.copy())
        self.t = np.array(t.copy())

#
# señales básicas
#

# escalón básico
def escalon(t):
    return sennal(1.0*(np.array(t) >= 0), t)

# impulso básico
def impulso(t):
    return sennal(1.0*(np.array(t) == 0), t)

# rampa
def rampa(t):
    return sennal(np.array(t), t)

# sinusoide
def sinusoide(t, omega, theta):
    return sennal(1.0*np.cos(omega*np.array(t)+theta), t)

#
# pruebas
#

# señales
t = np.arange(-5, 10+1)

s1 = impulso(t)
plot_stem(s1.t, s1.d, title="impulso")

s2 = escalon(t)
plot_stem(s2.t, s2.d, title="escalón")

s3 = rampa(t)
plot_stem(s3.t, s3.d, title="rampa")

s4 = sinusoide(t, np.pi/5, np.pi/3)
plot_stem(s4.t, s4.d, title="sinusoide")
```

```python
#
# bloques básicos
#

# sumador de señales
def sumador(s1, s2):
    if not isinstance(s1, sennal): raise Exception("only signals, please")
    if not isinstance(s2, sennal): raise Exception("only signals, please")
    # revisar el boletín 2
    tt = np.arange(min(s1.t[0], s2.t[0]), max(s1.t[-1], s2.t[-1])+1)
    dd = []
    for i in tt:
        v = 0.0
        if i>=s1.t[0] and i<=s1.t[-1]:
            v = v + s1.d[i-s1.t[0]]
        if i>=s2.t[0] and i<=s2.t[-1]:
            v = v + s2.d[i-s2.t[0]]
        dd.append(v)
    return sennal(dd, tt)

# restador
def restador(s1, s2):
    if not isinstance(s1, sennal): raise Exception("only signals, please")
    if not isinstance(s2, sennal): raise Exception("only signals, please")
    return sumador(s1, sennal(-1.0 * s2.d, s2.t))

# multiplicador de señales
def multiplicador(s1, s2):
    if not isinstance(s1, sennal): raise Exception("only signals, please")
    if not isinstance(s2, sennal): raise Exception("only signals, please")
    # revisar el boletín 2
    tt = np.arange(min(s1.t[0], s2.t[0]), max(s1.t[-1], s2.t[-1])+1)
    dd = []
    for i in tt:
        if i>=s1.t[0] and i<=s1.t[-1] and i>=s2.t[0] and i<=s2.t[-1]:
            dd.append(s1.d[i-s1.t[0]] * s2.d[i-s2.t[0]])
        else:
            dd.append(0.0)
    return sennal(dd, tt)

# escalador de señal
def escalador(s, n=1.0):
    if not isinstance(s, sennal): raise Exception("only signals, please")
    if not isinstance(float(n), float): raise Exception("should be a number, please")
    return sennal(n * s.d, s.t)

# retardador de señal
def retardador(s, n=0):
    if not isinstance(s, sennal): raise Exception("only signals, please")
    if not isinstance(n, int): raise Exception("should be an integer, please")
    if n < 0: raise Exception("should be a positive integer, please")
    dd = np.insert(s.d.copy(), 0, np.zeros(n))
    tt = np.append(s.t.copy(), np.arange(s.t[-1] + 1, s.t[-1] + 1 + n))
    return sennal(dd, tt)

# retardador de señal con pérdida de datos
# POR PETICIÓN POPULAR, un retardador que no aumenta el tamaño de la señal :)
def retardador_con_perdidas(s, n=0):
    if not isinstance(s, sennal): raise Exception("only signals, please")
    if not isinstance(n, int): raise Exception("should be an integer, please")
    if n < 0: raise Exception("should be a positive integer, please")
    if n >= len(s.d):
        return sennal(np.zeros(len(s.d)), s.t)
    else:
        dd = np.insert(s.d.copy(), 0, np.zeros(n))
        return sennal(dd[:len(s.d)], s.t)

# inversor de señal
def inversor(s):
    if not isinstance(s, sennal): raise Exception("only signals, please")
    dd = [1/i if i!=0 else 0 for i in s.d]
    return sennal(dd, s.t)

#
# pruebas
#

t = np.arange(-10, 10+1)

s1 = impulso(t)
s2 = escalon(t)
s3 = rampa(t)

# sumador
s = sumador(s1, s2)
plot_stem(s.t, s.d, title="sumador")

# multiplicador
s = multiplicador(s2, s3)
plot_stem(s.t, s.d, title="multiplicador")

# escalador
s = escalador(s3, -2)
plot_stem(s.t, s.d, title="escalador")

# retardador
s = retardador(s2, 4)
plot_stem(s.t, s.d, title="retardador")

# retardador con pérdidas
#s = retardador_con_perdidas(s2, 4)
#plot_stem(s.t, s.d, title="retardador con pérdidas")

# inversor
s = inversor(s3)
plot_stem(s.t, s.d, title="inversor")
```

```python
#
# circuitos básicos
#

#
# primera versión: cálculo punto a punto
#

# media móvil

#  n |   x  | y
# ----------------
# -1 | x[-1]| y[-1]
#  0 | x[0] | y[0] = (x[0] + x[-1]) * 0.5
#  1 | x[1] | y[1] = (x[1] + x[0]) * 0.5
#  2 | x[2] | y[2] = (x[2] + x[1]) * 0.5
#  3 | x[3] | y[3] = (x[3] + x[2]) * 0.5
# ...

def circuito1(x):
    dd = []
    x_retardado = 0
    for x_actual in x.d:
        dd.append((x_actual + x_retardado) * 0.5)
        x_retardado = x_actual
    return sennal(dd, x.t)

#s1 = impulso(np.arange(-2, 8+1))
#s2 = circuito1(s1)
#plot_stem(s1.t, s1.d, title="señal original")
#plot_stem(s2.t, s2.d, title="señal procesada")

s3 = escalon(np.arange(-2, 8+1))
s4 = circuito1(s3)
plot_stem(s3.t, s3.d, title="señal original")
plot_stem(s4.t, s4.d, title="señal procesada")
print("señal original: {}".format(s3.d))
print("señal procesada: {}".format(s4.d))

#s5 = sennal([6, 8, 4, 6, 0, 7, 5, 10, 10, 5], np.arange(0, 10))
#s6 = circuito1(s5)
#plot_stem(s5.t, s5.d, title="señal original")
#plot_stem(s6.t, s6.d, title="señal procesada")
```

```python
#
# circuitos básicos
#

#
# primera versión: cálculo punto a punto
#

# diagrama de bloques con salida retardada

#  n |   x  | y
# ----------------
# -1 | x[-1]| y[-1]
#  0 | x[0] | y[0] = (x[0] + x[-1]) * 0.5 + y[-1] * 0.25
#  1 | x[1] | y[1] = (x[1] + x[0]) * 0.5 + y[0] * 0.25
#  2 | x[2] | y[2] = (x[2] + x[1]) * 0.5 + y[1] * 0.25
#  3 | x[3] | y[3] = (x[3] + x[2]) * 0.5 + y[2] * 0.25
# ...

def circuito2(x):
    dd = []
    x_retardado = 0
    y_retardado = 0
    for x_actual in x.d:
        y_actual = (x_actual + x_retardado) * 0.5 + y_retardado * 0.25
        dd.append(y_actual)
        x_retardado = x_actual
        y_retardado = y_actual
    return sennal(dd, x.t)

#s1 = impulso(np.arange(-2, 8+1))
#s2 = circuito2(s1)
#plot_stem(s1.t, s1.d, title="señal original")
#plot_stem(s2.t, s2.d, title="señal procesada")

#s3 = escalon(np.arange(-2, 8+1))
#s4 = circuito2(s3)
#plot_stem(s3.t, s3.d, title="señal original")
#plot_stem(s4.t, s4.d, title="señal procesada")

s5 = sennal([6, 8, 4, 6, 0, 7, 5, 10, 10, 5], np.arange(0, 10))
s6 = circuito2(s5)
plot_stem(s5.t, s5.d, title="señal original")
plot_stem(s6.t, s6.d, title="señal procesada")
print("señal original: {}".format(s5.d))
print("señal procesada: {}".format(s6.d))
```

```python
#
# circuitos básicos
#

#
# primera versión: cálculo punto a punto
#

# diagrama de bloques para la raíz cuadrada

#  n |   x  | y
# ----------------
# -1 | x[-1]| y[-1]
#  0 | x[0] | y[0] = ((x[0] * 1/y[-1]) + y[-1]) * 0.5
#  1 | x[1] | y[1] = ((x[1] * 1/y[0]) + y[0]) * 0.5
#  2 | x[2] | y[2] = ((x[2] * 1/y[1]) + y[1]) * 0.5
#  3 | x[3] | y[3] = ((x[3] * 1/y[2]) + y[2]) * 0.5
# ...

def circuito3(x, valor_aproximado=1):
    dd = []
    x_retardado = 0
    y_retardado = valor_aproximado
    for x_actual in x.d:
        y_actual = ((x_actual/y_retardado) + y_retardado) * 0.5
        dd.append(y_actual)
        x_retardado = x_actual
        y_retardado = y_actual
    return sennal(dd, x.t)

#s1 = impulso(np.arange(-2, 8+1))
#s2 = circuito2(s1)
#plot_stem(s1.t, s1.d, title="señal original")
#plot_stem(s2.t, s2.d, title="señal procesada")

#s3 = escalon(np.arange(-2, 8+1))
#s4 = circuito2(s3)
#plot_stem(s3.t, s3.d, title="señal original")
#plot_stem(s4.t, s4.d, title="señal procesada")

#s5 = sennal([6, 8, 4, 6, 0, 7, 5, 10, 10, 5], np.arange(0, 10))
#s6 = circuito3(s5)
#plot_stem(s5.t, s5.d, title="señal original")
#plot_stem(s6.t, s6.d, title="señal procesada")

s7 = sennal([9]*20, np.arange(0, 20))
s8 = circuito3(s7, 21)
#s8 = circuito3(s7)
plot_stem(s7.t, s7.d, title="señal original")
plot_stem(s8.t, s8.d, title="señal procesada")
print("señal original: {}".format(s7.d))
print("señal procesada: {}".format(s8.d))
```

```python
#
# circuitos básicos
#

#
# segunda versión: cálculo directo como señal
#

# comentario:
#
# por regla general, los bloques básicos que
# hemos definido y que trabajan a nivel de
# señal sólo resultarán útiles mientras no haya
# realimentación desde la salida
#
# cuando la hay, hay que iterar continuamente
# para ir reaprovechando resultados anteriores

# en este circuito tenemos suerte y no hay realimentación

# comentario:
#
# prueba a sustituir
#     retardador(x, 1)
# por
#     retardador_con_perdidas(x, 1)

# media móvil

def circuito1_v2(x):
    return escalador(sumador(x, retardador(x, 1)), 0.5)

s1 = escalon(np.arange(-2, 8+1))
s2 = circuito1_v2(s1)
plot_stem(s1.t, s1.d, title="señal original")
plot_stem(s2.t, s2.d, title="señal procesada")
print("señal original: {}".format(s1.d))
print("señal procesada: {}".format(s2.d))
```

```python
# la mayor ventaja de este formato es que podemos extenderlo sin demasiados problemas

# imagina que queremos hacer la media de N números
# en lugar de sólo 2 números:
#     y[n] = (x[n] + x[n-1] + x[n-2] + ... + x[n-(N-1)])/N

def circuito1_v2_extendido(x, n=2):
    if n < 1: raise Exception("ops! value n must be a positive integer")
    y = sennal(x.d, x.t)
    for i in range(1, n):
        y = sumador(y, retardador(x, i))
    return escalador(y, 1/n)

s1 = escalon(np.arange(-2, 8+1))
s2 = circuito1_v2_extendido(s1, 4)
plot_stem(s1.t, s1.d, title="señal original")
plot_stem(s2.t, s2.d, title="señal procesada")
print("señal original: {}".format(s1.d))
print("señal procesada: {}".format(s2.d))
```

```python
#
# circuitos básicos
#

#
# segunda versión: cálculo directo como señal
#

# diagrama de bloques con salida retardada

# cuidado! en este circuito tenemos realimentación!

def circuito2_v2(x):
    y = sennal(np.zeros(len(x.d)), x.t.copy())
    # en cada iteración usamos una señal y[] mejor corregida
    # para calcular únicamente un nuevo valor y[i]
    for i in range(len(x.d)):
        #
        # recalcular con los bloques básicos
        #
        y = sumador(escalador(sumador(x, retardador(x, 1)), 0.5), escalador(retardador(y, 1), 0.25))
    return sennal(y.d[:len(x.d)], y.t[:len(x.d)])

s1 = sennal([6, 8, 4, 6, 0, 7, 5, 10, 10, 5], np.arange(0, 10))
s2 = circuito2_v2(s1)
print("señal original: {}".format(s1.d))
print("señal procesada: {}".format(s2.d))
```

```python
#
# circuitos básicos
#

#
# segunda versión: cálculo directo como señal
#

# diagrama de bloques para la raíz cuadrada

# cuidado! en este circuito tenemos realimentación!

def circuito3_v2(x, valor_inicial=1):
    y = sennal(np.zeros(len(x.d)+1), np.insert(x.t.copy(), 0, x.t[0]-1))
    y.d[0] = valor_inicial
    # en cada iteración usamos una señal y[] mejor corregida
    # para calcular únicamente un nuevo valor y[i]
    for i in range(len(x.d)):
        #
        # recalcular con los bloques básicos
        #
        y0 = retardador(y, 1)
        r0 = inversor(y0)
        r1 = multiplicador(x, r0)
        r2 = sumador(r1, y0)
        y = escalador(r2, 0.5)
        print("debug: el valor actual es {}".format(y.d[i+1]))
    return sennal(y.d[:len(x.d)+1], y.t[:len(x.d)+1])

s1 = sennal([9]*10, np.arange(0, 10))
s2 = circuito3_v2(s1, 16)
print("señal original: {}".format(s1.d))
print("señal procesada: {}".format(s2.d))
```

```python
#
# puedes comprobar que circuito2_v2() y circuito3_v2 tienen una misma estructura:
#
# - inicializar y[]
# - iterar sobre x:
#     - operar con bloques
# - devolver la primera parte de la señal y[] calculada
#
# podemos convertir circuito1() al mismo formato
#

def circuito1_v3(x):
    y = sennal(np.zeros(len(x.d)), x.t.copy())
    # en cada iteración usamos una señal y[] mejor corregida
    # para calcular únicamente un nuevo valor y[i]
    for i in range(len(x.t)):
        #
        # recalcular con los bloques básicos
        #
        y = escalador(sumador(x, retardador(x, 1)), 0.5)
    return sennal(y.d[:len(x.d)], y.t[:len(x.d)])

s1 = escalon(np.arange(-2, 8+1))
s2 = circuito1_v3(s1)
print("señal original: {}".format(s1.d))
print("señal procesada: {}".format(s2.d))
```
