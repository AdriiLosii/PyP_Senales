# Diagramas de bloques
* Representan circuitos para el análisis y síntesis de las señales
*  Nos permiten diseñar de forma abstracta algoritmos sin importar como serán implementados
* Se realizar empleando  bloques de construcción básicos que representan las operaciones más elementales sobre las señales
* El funcionamiento resumido es el siguiente:
 * El circuitor/sistema recibe una secuencia de entrada $x[n]$
 * Se aplican a las operaciones definidas en los bloques de construcción que componen el circuito
 * El circuito/sistema devuelve una secuencia de salida $y[n]$
 
## Objetivo de la práctica
En esta práctica vamos a implementar los bloques básicos vistos en la teoría y con ellos codificaremos varios diagramas de bloques. Realizar nuestra propia implementación nos ayudará a fijar conceptos y entender mejor las operaciones y ciclos de los diagramas de bloques

Salvando las distancias, puesto que esta práctica es introductoria, la idea es tener una librería de bloques básicos que podamos ir extendiendo para crear nuestros circuitos, al estilo de dsptools (librería que proporciona el MIT) en http://web.mit.edu/6.02/www/s2009/dsptools/index.html

```python
#librerias necesarias para la práctica
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from random import randrange
import IPython
plt.rcParams["figure.figsize"] = (14,4) #dimensiones por defecto al plotear (width, height)

#Este es un método de visualización muy básico para hacer pruebas
def plot_stem( t , y, y_lim=None, view_baseline=True, title=None):
    """Función que configura y visualiza un gráfico stem 

       Parámetros:
       t -- muestras de tiempo
       y -- secuencia para visualizar
       y_lim -- tupla de la forma (min, max) que establece los límites en el eje de ordenadas. Por defecto None
       view_baseline -- Establece una línea horizontal en el eje de las x a la altura de  y=0 . Por defecto False

    """
    markerline, stemlines, baseline = plt.stem(t, y)
    plt.setp(baseline, visible=view_baseline)
    plt.setp(markerline, color="crimson",markersize = 10)
    plt.setp(stemlines, color="crimson", linewidth = 4)
    plt.axvline(x=0, linewidth=1,color="black")
    plt.ylim(y_lim)
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
class signal():
    def __init__(self, d, t):
        self.d = np.array(d)
        self.t = np.array(t)


# Señales básicas:
def escalon(t):
    return signal(1*(t>=0), t)

def impulso(t):
    return signal(1*(t==0), t)

def rampa(t):
    return signal(t*(t>=0), t)

def sinusoide(t, omega, theta):
    return signal(1.0*np.cos(omega*np.array(t)+theta), t)

# Bloques básicos:
def sumador(s1, s2):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")
    if not isinstance(s2, signal): raise Exception("Error, solo señales")

    add_interv = np.arange(max(min(s1.t),min(s2.t)), min(max(s1.t),max(s2.t))+1, 1)
    add_signal = []

    for i in range(len(add_interv)):
        add_signal.append(s1.d[i]+s2.d[i])

    return signal(add_signal, add_interv)

def multiplicador(s1, s2):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")
    if not isinstance(s2, signal): raise Exception("Error, solo señales")

    add_interv = np.arange(max(min(s1.t),min(s2.t)), min(max(s1.t),max(s2.t))+1, 1)
    add_signal = []

    for i in range(len(add_interv)):
        add_signal.append(s1.d[i]*s2.d[i])

    return signal(add_signal, add_interv)

def escalador(s, n):
    if not isinstance(s, signal): raise Exception("Error, solo señales")
        
    return signal(s.d*n, s.t)

def retardador(s, n):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")
    if not isinstance(n, int): raise Exception("Error, solo enteros")

    new_signal = []

    if(n==0):
        return signal(s.d, s.t)
    else:
        for i in range(n):
            new_signal.append(0)

        for i in range(len(s.d)-n):
            new_signal.append(s.d[i])

        return signal(new_signal, s.t)

def inversor(s):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")

    new_signal = []
    for i in range(len(s.d)):
        if(s.d[i]!=0):
            new_signal.append(1/s.d[i])
        else:
            new_signal.append(s.d[i])

    return signal(new_signal, s.t)


# Pruebas:
t = np.arange(-5, 10+1)

# Señales básicas:
s1 = impulso(t)
plot_stem(s1.t, s1.d, title="Impulso")

s2 = escalon(t)
plot_stem(s2.t, s2.d, title="Escalón")

s3 = rampa(t)
plot_stem(s3.t, s3.d, title="Rampa")

s4 = sinusoide(t, np.pi/5, np.pi/3)
plot_stem(s4.t, s4.d, title="Sinusoide")

# Operaciones básicas:
sumador = sumador(s1, s2)
plot_stem(sumador.t, sumador.d, title="Sumador")

escalador = escalador(s3, -1)
plot_stem(escalador.t, escalador.d, title="Escalador")

retardador = retardador(s1, 3)
plot_stem(retardador.t, retardador.d, title="Retardador")

inversor = inversor(s3)
plot_stem(inversor.t, inversor.d, title="Inversor")
```

### Cálculo de una raíz cuadrada 
Muchas computadoras y calculadoras calculan la raíz cuadrada de un número positivo $A$ aplicando el algoritmo iterativo que representa el siguiente diagrama 

![title](raiz_cuadrada.png)

Si la secuencia de entrada al sistema es un escalón unidad de amplitud A (es decir, $x[n]=Au[n]$) y utilizamos como  condición inicial $y[-1]$ un valor estimado de  $\sqrt A$, la respuesta del sistema tenderá a $\sqrt A$ cuando $n$ aumente.


<b>Nota</b>: para que este diagrama funcione como queremos tenemos que modificar levemente el bloque que almacena los retardos para permitirle tener un valor inicial diferente de cero. Podemos inicializarlo a un valor concreto a través del constructor.

<b>Nota2</b>: No hace falta que $y[-1]$ se inicialice a una buena estimación de $\sqrt A$ pero, obviamente, si es mas exacta convergerá antes. Podéis probar inicializando a 1 y ver que pasa




```python
# Bloques básicos:
def sumador(s1, s2):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")
    if not isinstance(s2, signal): raise Exception("Error, solo señales")

    add_interv = np.arange(max(min(s1.t),min(s2.t)), min(max(s1.t),max(s2.t))+1, 1)
    add_signal = []

    for i in range(len(add_interv)):
        add_signal.append(s1.d[i]+s2.d[i])

    return signal(add_signal, add_interv)

def multiplicador(s1, s2):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")
    if not isinstance(s2, signal): raise Exception("Error, solo señales")

    add_interv = np.arange(max(min(s1.t),min(s2.t)), min(max(s1.t),max(s2.t))+1, 1)
    add_signal = []

    for i in range(len(add_interv)):
        add_signal.append(s1.d[i]*s2.d[i])

    return signal(add_signal, add_interv)

def escalador(s, n):
    if not isinstance(s, signal): raise Exception("Error, solo señales")
        
    return signal(s.d*n, s.t)

def retardador(s, n):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")
    if not isinstance(n, int): raise Exception("Error, solo enteros")

    new_signal = []

    if(n==0):
        return signal(s.d, s.t)
    else:
        for i in range(n):
            new_signal.append(0)

        for i in range(len(s.d)-n):
            new_signal.append(s.d[i])

        return signal(new_signal, s.t)

def inversor(s):
    if not isinstance(s1, signal): raise Exception("Error, solo señales")

    new_signal = []
    for i in range(len(s.d)):
        if(s.d[i]!=0):
            new_signal.append(1/s.d[i])
        else:
            new_signal.append(s.d[i])

    return signal(new_signal, s.t)


# Circuitos:
def circuito1(x):
    #Una forma:
    """
    x_retardado = retardador(x, 1)
    y1 = sumador(x, x_retardado)
    y = escalador(y1, 0.5)
    """
    
    #Otra forma:
    y = escalador(sumador(x, retardador(x, 1)), 0.5)

    return y

def circuito2(x):
    #Una forma:
    """
    y = escalador(sumador(x, retardador(x, 1)), 0.5)
    y1 = sumador(y, retardador(y, 1))
    """

    #Otra forma:
    y = circuito1(x)
    y1 = sumador(y, escalador(retardador(y, 1), 0.25))

    return y1

def circuito3(x, valor_aprox=1):
    y_0 = escalador(x, 1/2)
    y = sumador(multiplicador(x, inversor(retardador(y_0, 1))), retardador(y_0, 1))

    return y


t = np.arange(-5, 10+1, 1)

s1 = escalon(t)
plot_stem(s1.t, s1.d, title="Señal original")

s2 = circuito1(s1)
plot_stem(s2.t, s2.d, title="Media móvil")

s3 = circuito2(s1)
plot_stem(s3.t, s3.d, title="Diagrama de bloques con salida retardada")

s4 = circuito3(s1, )
plot_stem(s4.t, s4.d, title="Cálculo de una raíz cuadrada")
```
