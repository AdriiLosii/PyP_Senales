## Análisis de temperaturas

En esta práctica vamos a trabajar con las temperaturas de Lugo diez-minutales de los años 2018-2020 (3 años). 

La idea es imaginarnos que tenemos que procesar datos reales y analizarlos en frecuencia para intentar sacar u observar alguna característica.




```python
#librerias necesarias para la práctica
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.rcParams["figure.figsize"] = (14,4) #dimensiones por defecto al plotear (width, height)
```

### Obtención de datos
Lo primero que tendríamos que hacer es ir a la web de Meteogalicia para descargar los datos de temperatura de la parte de [históricos](https://www.meteogalicia.gal/observacion/estacionshistorico/historico.action?idEst=10050) (yo ya lo hice por vosotros). En particular, el fichero que os proporciono tiene datos de la temperatura media a 1.5 metros cada 10 minutos de los años 2018, 2019, 2020.

Vamos a cargar los datos empleando [Pandas](https://pandas.pydata.org/),  una de las librerías más empleadas para la analítica de datos con Python. Un DataFrame de Pandas es algo parecido a una hoja de Excel en la que tenemos filas y columnas y una clave que identifica cada fila. Sobre las filas y columnas se pueden hacer operaciones de resumen, analítica, filtrado, etc.

#### Carga de datos

Pandas permite cargar datos con facilidad desde diferentes soportes. Uno de los métodos es la carga de ficheros CSV ([read_csv](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html)). Durante este proceso se pueden emplear diferentes argumentos que indican cómo se hará el proceso como, por ejemplo, el tipo de separador que se usa, si tiene que "escapar" una serie de filas, columnas que se quieren cargar, valor para identificar los nulos, etc.

```python
#Meteogalicia usa el valor -9999 para los NaN y por eso lo configuro al cargar los datos
datos=pd.read_csv("temperaturas_lugo.csv", na_values=[-9999, -9999.0])
#Cubro los huecos generados por los NaN directamente en el dataframe de datos
#<<pad>> sirve para emplear la última medida válida para cubrir los huecos
#<<inplace>> modifica internamente el dataframe, en caso contrario devolvería una copia
datos.fillna(inplace=True, method='pad')
#Visualizo las primeros 'n' filas. n=5 por defecto
datos.head()

```

#### Gestión de fechas con Pandas
Al cargar los datos desde CSVs no se identifican correctamente las fechas y se clasifican como "objects" por Pandas. Pandas tiene mucha potencia trabajando con fechas por lo que vamos a informar que el campo identificado como *fecha* es, efectivamente, una fecha ([DateTime](https://docs.python.org/3/library/datetime.html#datetime.datetime)). 
El método [to_datetime()](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) nos permitirá hacer ese cambio. Ese método nos permite informale del tipo de formato que se va a usar e incluso la zona horaria empleada.

Además, vamos a configurar Pandas para que use el campo "fecha" como clave ([set_index](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html)), ya que no se repiten y eso facilita luego las operaciones sobre los datos.

```python
#Por defecto Pandas cargó el campo "fecha" como un object en lugar de como una fecha.
datos.dtypes
```

```python
#Convertimos la fecha a DateTime
datos["fecha"]=pd.to_datetime(datos['fecha'])
#Empleo el campo fecha como clave
datos.set_index("fecha", inplace=True)

#Visualizamos la cabecera
datos.head()

```

#### Resample de las fechas
Trabajar cada 10 minutos es una resolución excesivamente pequeña e incomoda para interpretar. Vamos a transformar los datos y emplear una resolución de 1 hora. para eso emplearemos la operación [resample](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html) de Pandas y haremos medias de los datos que caen en cada hora. Para poder realizar esto, es necesario que el index del dataframe sea una fecha. el método [fillna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html) nos permite cubre huecos (NaN) que se generen por falta de datos al hacer la operación de resumen (mean).



```python
#Particularmente me parece una buena costumbre no "sobrescribir el dataframe original 
#con una operación de resumen y generar una copia para mantener los nuevos datos
datos_resample=datos.resample("1H").mean().fillna(method='pad')

```

#### Visualización de datos
Matplotlib es bastante tedioso para visualizar, existen librerías más cómodas a la hora de hacer visualizaciones científicas. Una interesante es [seaborn](https://seaborn.pydata.org/)

```python
#Podemos visualizar los datos con alguna librería más cómoda que matplotlib como, por ejemplo, seaborn
sns.lineplot(data=datos_resample)
```
