# Análisis de Frecuencia III


En esta práctica vamos a trabajar con la **Transformada Discreta de Fourier** (DFT)  y vamos a ver como aplicarla de forma práctica. 

### Introducción

Tal y como hemos comentado en prácticas anteriores, lo habitual es trabajar con señales finitas y discretas en el tiempo y usar algún tipo de procesador digital de señales (ej. computador digital) para analizarlas.

En las clases de teoría vimos que la forma de realizar el análisis en frecuencia de una señal discreta en el tiempo, $x[n]$,  es a través de la transformada de fourier $X(\omega)$. La dificultad que nos surge es que $X(\omega)$ es una función continua y, por tanto, no es una representación adecuada computacionalmente (computacionalmente lo adecuado es trabajar en discreto).

La DFT nos permite realizar una representación de $x[n]$ en el dominio de la frecuencia a través de muestras del espectro de $X(\omega)$, es decir, tendríamos a la entrada y a la salida secuencias discretas.

Desde un punto de vista conceptual, la idea es la siguiente:


Sabemos que podemos desarrollar la serie de Fourier sobre una señal discreta y periódica en el dominio del tiempo y así obtener una secuencia (discreta) de coeficientes de Fourier ($c_k$), que forman una secuencia periódica de período N (el mismo período que la señal en el dominio del tiempo). Si creamos una señal periódica partiendo de nuestra señal finita y discreta, entonces podríamos desarrollar una serie de Fourier sobre ella. Los coeficientes que obtendríamos ($c_k$) representarían una secuencia de muestras de la transformada de Fourier.

Para generar la señal periódica ($x_p[n]$) partiendo de $x[n]$, necesitamos **establecer un período N cuyo valor sea mayor o igual que la longitud de la señal $x$** que estamos analizando y repetir la secuencia en función del período seleccionado.

$\large x_p[n]=\sum\limits_{r=-\infty}^{\infty}x[n-rN]$



Si establecemos un N>len(x) deberemos rellenar con ceros los valores de la secuencia *nuevos*. Recuperar la señal original desde señal la periódica es sencillo:


$\large  x[n] = \left \{ \begin{matrix} x_p[n] , & 0<=n<=N-1
\\ 0 & en~otro~caso\end{matrix}\right.  $

Una vez que *creamos* nuestra señal periódica, podremos desarrollar la DFT y su inversa a través de las siguientes ecuaciones:

**DFT** - Ecuación de análisis o transformada directa


$\large  X[k]=\sum\limits_{n=0}^{N-1} x[n]e^{-j2\pi kn/N}, 0<=k<=N-1$

**IDFT** - Ecuación de síntesis o transformada inversa

$\large  x[n]=\frac{1}{N}\sum\limits_{n=0}^{N-1} X[k]e^{j2\pi kn/N}, 0<=n<=N-1$


## Importancia de la DFT

Además de su importancia teórica, la DFT tiene un papel crucial en la realización de algoritmos de procesamiento digital de señales por una razón fundamental: existen algoritmos muy eficientes para su cálculo. Esto lo convierte en una herramienta muy potente para el análisis de señales en el dominio de la frecuencia.

Típicamente la DFT se calcula a través de algoritmos de la familia FFT (Transformada Rápida de Fourier). Podemos encontrar algoritmos FFT en muchos paquetes, entre ellos en [numpy](https://numpy.org/doc/stable/reference/routines.fft.html) y [scipy](https://docs.scipy.org/doc/scipy/reference/fft.html#module-scipy.fft). En esta práctica emplearemos numpy pero ambos paquetes son muy similares (scipy tiene un conjunto más amplio de rutinas).



Lo primero es calcular la DFT que en numpy se hace a través de una algoritmo de la familia de *Fast Fourier Transform* (FFT)

<code>numpy.fft.fft(signal,N)</code>

El valor N representa el período que vamos a emplear para desarrollar la DFT y que tiene que cumplir lo siguiente 

$N>=len(x)$

El resultado de la FFT, es la transformada $X[k]$ en la que cada elemento es un complejo con su módulo y su fase (un *numpy.ndarray* de complejos). Lo típico es visualizar el espectro de módulo (la magnitud o módulo de los valores obtenidos a través de la FFT).

```python
X=np.fft.fft(x)#La función devuelve un ndarray de complejos
print(type(X))
print(type(X[0]))

#Imprimimos varios elementos de la DFT para ver que son complejos (tienen parte real e imaginaria)
print(X[:5])

print("parte real")
print(X[:5].real)
print("parte imaginaria")
print(X[:5].imag)

print("Módulo")
print(abs(X[:5]))


plt.stem(abs(X))#abs para obtener el módulo

plt.show()


```

## Frecuencias positivas y negativas

Recordad que los coeficientes de la serie son periódicos de período $2\pi$. Solo necesitamos un período continuo para obtener todos los posibles valores. Los valores obtenidos a través de la FFT cubren el período $[0,2\pi]$ pero en muchas ocasiones trabajamos en el período $[-\pi,\pi]$. Si queremos pasar de un punto a otro tenemos que reinterpretar las posiciones del array resultante.



Los valores obtenidos y la gráfica anterior son el resultado de aplicar directamente la FFT pero es necesario reinterpretarlos si queremos posicionarlos en el período $[-\pi, \pi]$, ya que por defecto siguen el denominado orden *standard*  que consiste en lo siguiente:

Si tenemos que X=fft(x,N) entonces:

* X[0]: tiene la frecuencia cero, que corresponde con la suma de la señal de entrada (Si k=0 entonces tenemos $e^0=1$ por eso el resultado es la suma de la señal). Este valor es siempre real para señales de entrada reales

* X[1:n//2]: contiene las frecuencias positivas ($k>0$)

* X[n//2+1:]: contiene las frecuencias negativas de menor a mayor, es decir, el $k=-1$ es el len(X)-1



El resultado de cambiar el orden en base a lo descrito anteriormente sería algo parecido a (secuencias impares):

<code>np.concatenate(X[n//2+1:],X[0:n//2])</code>



Debido a la periocidad de período $2\pi$  que tenemos en la transformada, sabemos que una frecuencia positiva $\omega$ que está entre $\pi$ y $2\pi$ es equivalente a la frecuencia negativa $\omega - 2\pi$ y por eso podemos intepretar que los índices de la mitad superior del array obtenido por la FFT son equivalentes a las frecuencias negativas.






Es muy importante que tengamos en cuenta el orden (período)  en el que nos dan los valores de la transformada para interpretarlos correctamente.


**Importante**: Existe una diferencia entre procesar señales con elementos pares o con elementos impares que tenéis que tener en cuenta. Si x tiene una longitud impar, $X$ es simétrico en torno al punto central $X[0]$. En el caso de señales de entrada con elementos pares, el resultado de la FFT no será simétrico pero 
tendrá un valor *compartido*.  La frecuencia positiva máxima será $X[N//2]$ que corresponde con $\omega_{N/2}=\pi$. Teniendo en cuenta las propiedades simétricas de la transformada, sabemos que las frecuencias de $-\pi$ y $\pi$ son iguales ($X[-k]=X[k]$), por lo que este valor será también empleado para la frecuencia negativa correspondiente. *Compartiendo* este valor, podemos tener una transformada simétrica también para longitudes pares.

## Reajuste de las frecuencias

El orden de los valores que obtenemos con la FFT de numpy es habitual y lo podemos encontrar en otros librerías y por eso desarrollamos una función para reordenarlo y entender cómo se hace pero, en  particular, numpy proporciona una función que permite hacerlo de forma automática: [numpy.fft.fftshift](https://numpy.org/doc/stable/reference/generated/numpy.fft.fftshift.html#numpy.fft.fftshift). Lo único que tenéis que tener en cuenta es que solo ordena los valores obtenidos a través de la FFT, es decir, **no devuelve un índice nuevo ni crea una secuencia simétrica en el caso de que x tenga longitud par**. 



**Importante:** Aunque nosotros queramos trabajar con una transformada centrada en el k=0, las funciones de numpy esperan el orden *standard* y por eso existe una función que permite volver al orden *standard*: [numpy.fft.ifftshift](https://numpy.org/doc/stable/reference/generated/numpy.fft.ifftshift.html#numpy.fft.ifftshift)




```python
#Reordenamos la DFT centrada en el 0
#Recordad que si len(x) es par, la DFT no es simétrica salvo que 
#le inserteis el valor de la frecuencia negativa máxima, tal y como se comenta anteriormente

X_shifted=np.fft.fftshift(X_par)


#X_shifted=np.append(X_shifted,[X_shifted[0]])##Inserto para conseguir simetria -pi:pi
#####Inicio: Esta parte es necesaria si queremos modificar el eje x#####
N=len(X_shifted)
extra=1 if N%2!=0 else 0
ordered_index=np.arange(-(N//2), N//2+extra)
#####Fin: Esta parte es necesaria si queremos modificar el eje x#####

plt.stem(ordered_index, abs(X_shifted))
plt.show()




##Teniendo en cuenta que las funciones de numpy esperan
#el orden standard, es bueno conocer que existe una función
#que permite recuperar el orden original
X_original=np.fft.ifftshift(X_shifted)
plt.stem(abs(X_original))
plt.show()
```

## Transformada inversa

Tal y como hemos visto al comienzo, la ecuación para la transformada inversa es la siguiente:

**IDFT** - Ecuación de síntesis o transformada inversa

$\large x[n]=\frac{1}{N}\sum\limits_{n=0}^{N-1} X[k]e^{j2\pi kn/N}, 0<=n<=N-1$

Esta ecuación nos permite procesar una DFT y obtener la secuencia original en el dominio del tiempo ($x[n]$).

Numpy tiene una función para realizar IDTF ([numpy.fft.ifft](https://numpy.org/doc/stable/reference/generated/numpy.fft.ifft.html#numpy.fft.ifft)), por lo que no necesitamos implementarla. **Ojo** que la función devuelve números complejos


<b>Importante:</b> La función de **IDTF de numpy espera una transformada en el orden *standard*** que genera numpy.fft.fft, por lo que si hemos modificado el orden tendremos que retrocederlo



```python
plt.stem(np.fft.ifft(np.fft.fft(x_impar)))#Saca un warning por los números complejos
plt.show()


plt.stem(np.fft.ifft(np.fft.fft(x_impar)).real)#Ploteo la parte real
plt.show()
```

## Conversión de los indices de la DFT a frecuencias interpretables



A la hora de analizar los resultados de la DFT es importante que podamos relacionar los índices $k$ con una frecuencia ($f$) en el *mundo real* (ciclos por unidad de muestreo, como, por ejemplo ciclos por segundo) para poder analizar e interpretar los resultados. Numpy proporciona una función que convierte los índices en frecuencias [fft.fftfreq](https://numpy.org/doc/stable/reference/generated/numpy.fft.fftfreq.html#numpy.fft.fftfreq)



<code>fft.fftfreq(n, d=1.0)</code>

* **n**: representa N, la longitud de $X[k]$
* **d**: representa la frecuencia de muestreo. Ej. 0.5 puede representar 2 muestras por segundo




**Importante**: Esta función genera frecuencias positivas y negativas, por lo que  se mostrarían centradas en el 0 (algo parecido a lo que estábamos haciendo antes con los valores de la transformada). Si visualizamos la transformada ayudados de los índices de las frecuencias, matplotlib ya lo interpreta correctamente y no necesitamos modificar el orden de la DFT (en el ejemplo lo entendereis mejor)

Vamos a ver como hacerlo con un ejemplo práctico que sigue los siguientes pasos:


* Creamos tres señales sinusoidales con distintas frecuencias
* Las muestreamos cada 0.25 segundos durante 60 segundos (240 muestras)
* Sumamos las señales y obtenemos una nueva
* Obtenemos la DFT de la señal resultante
* Cambiamos los índices por las frecuencias empleando *fft.fftfreq*











```python

#nuestras muestras obtenidas cada 0.25 segundos
f_muestreo=1/4
N=60
t=np.linspace(0,N,int(N*(1/f_muestreo)))

#Creamos varias señales sinusoidales con distintas frecuencias
f1=1/60
f2=1/5
f3=1/2
data1=np.sin(2*np.pi*f1*t)
data2=np.sin(2*np.pi*f2*t)
data3=np.sin(2*np.pi*f3*t) 


#Sumamos y generamos una nueva señal compuesta de las 3
data=data1+data2+data3
plt.stem(data)
plt.show()

```

```python
#Generamos la DFT  y transformamos sus índices en una lista ordenada de las frecuencias
#Fijaos que ya no necesito reordenar los valores de la transformada para visualizarlos,
# ya que matplotlib los pinta en relación a la lista de índices pasados
#Al utilizar fftfreq me devuelve unos índices negativos y otros positivos por lo que
#se puede ordenar la visualización de índice menor a mayor

X_data=np.fft.fft(data)
X_freq=np.fft.fftfreq(len(X_data), f_muestreo)
plt.stem(X_freq, abs(X_data))

plt.show()

#podemos visualizar solo las frecuencias positivas
idx_pos=X_freq>0#Devuelve True o False según sea cierto
#ploteo solo los valores del índice que están a True
#abs es por ser complejo
plt.stem(X_freq[idx_pos], abs(X_data [idx_pos]))

plt.show()
```

Analizando la señal resultante podríamos ver que la mayoría de los valores están en torno al cero (ruido) pero que aparecen 3 picos importantes que nos dan información sobre la composición de la señal original.

Recordad que los coeficientes son un *valor de lo parecida* que es es una señal sinusoidal (a una frecuencia específica) con la señal original. Teniendo eso en mente, y viendo los resultados, podemos  suponer que esa señal original está compuesta principalmente por 3 señales (lo que es cierto). Otra interpretación es que a ciertas frecuencias ocurre *algo* que modifica la señal original.
